```python
# Loading data from the prediction.csv and displaying the first few rows of the DataFrame
import pandas as pd

df = pd.read_csv('prediction.csv')
print(df.head())
```

```python
# Exploring the dataset by checking its shape, info, descriptive statistics, and missing values


print("Shape of the DataFrame:",df.shape)

print("\nInformation about the DataFrame:")
df.info()

print("\nDescriptive statistics of the DataFrame:")
df.describe()

print("Number of the missing values in each column:")
df.isnull().sum()

```

```python
print("TARGET VARIABLE: Attrition")

print("\nHow many employees left vs stayed?")
print(df['Attrition'].value_counts())

print("\nPercentage")
print(df['Attrition'].value_counts(normalize=True)*100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set nice style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)

# Create bar chart
plt.figure()
df['Attrition'].value_counts().plot(kind='bar', color=['#2ecc71', '#e74c3c'])
plt.title('Attrition Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Attrition', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

# Add count labels on bars
for i, v in enumerate(df['Attrition'].value_counts()):
    plt.text(i, v + 20, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
